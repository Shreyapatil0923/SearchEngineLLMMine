```python
#############Arxiv----Research

####tools creation


from langchain_community.tools import ArxivQueryRun,wikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
```

```python
#######use the inbuild tool of wikipedia

api_wrapper_wiki=WikipediaAPIWrapper(top_k_results=1,doc_content_chars_maz=250)

wiki=wikipediaQueryRun(api_wrapper=api_wrapper_wiki)
wiki.name
```

```python
api_wrapper_arxiv=ArxivAPIWrapper(top_k_results=1,doc_content_chars_max=250)
arxiv=ArxivQueryRun(api_wrapper=api_wrapper_arxiv)

```

```python
tools=[wiki,arxiv]
```

```python
######custom tools[RAG Tool]

from langchain_community.document_loaders import  WebBaseLoader
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter


```

```python
loader=WebBaseLoader("https://")
docs=loader.load()
documents=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlaap=200).split_documents
vectordb=FAISS.from_documents(documents,OpenAIEmbeddings())
retriver=vectordb.as_retriever()
retriver

```

```python
from langchain.tools.retriver import create_retriver_tool
retriver_tool=create_retriver_tool(retriver,"langsmith-search","Search any information about langsmith")

retriver_tool.name


```

```python
tools=[wiki,arxiv,retriver_tool]
```
